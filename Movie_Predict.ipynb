```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv("tmdb_5000_movies.csv")
movies.head(3)
```

```python
credits=pd.read_csv("tmdb_5000_credits.csv")
credits.head(5)
```

```python
movies=movies.merge(credits,on='title')
movies.head(5)
```

```python
movies=movies[['movie_id',"title","overview","genres","keywords",'cast','crew','vote_average']]
movies.head(5)
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.iloc[0].genres
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
import ast
```

```python
movies['genres']=movies['genres'].apply(convert)
movies.head(4)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
movies.head(4)
```

```python
def convert3(obj):
    co=0
    L=[]
    for i in ast.literal_eval(obj):
        if co!=3:
            L.append(i['name'])
            co+=1
        else: 
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
movies.head(3)
```

```python
def dir(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if(i['job']=='Director'):
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(dir)
movies.head(3)
```

```python
movies['overview'][0]
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
movies.head(2)
```

```python
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(2)
```

```python
movies['tags']=movies['crew']+movies['cast']+movies['keywords']+movies['genres']+movies['overview']
movies.head(2)
```

```python
new_df=movies[['movie_id','title','tags','vote_average']]
new_df.head(3)
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
new_df.head(2)
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head(2)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df.head(3)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
new_df['title']=new_df['title'].apply(lambda x:x.lower())
```

```python
def rec(movie):
    movie=(lambda x:x.lower())(movie)
    movies_idx= new_df[new_df['title']==movie].index[0]
    distances=similarity[movies_idx]
    mov_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in mov_list:
        print(new_df.iloc[i[0]].title)
```

```python
rec('Batman Begins')
```
